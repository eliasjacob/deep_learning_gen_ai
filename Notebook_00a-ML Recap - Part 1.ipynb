## The Debate Between Two Cultures in Computational Modeling and Data Analysis

In the field of computational modeling and data analysis, a significant intellectual debate has emerged, centering on two distinct approaches:

1. Modeling Fundamental Mechanisms
2. Using Machine Learning for Prediction

This debate reflects a broader discussion about the most effective ways to understand and analyze complex phenomena. Let's explore these approaches in detail, along with a third emerging approach that combines elements of both.

### 1. Modeling the Fundamental Mechanism

This approach focuses on understanding phenomena at their key by simulating or representing the processes that lead to their occurrence. It is deeply rooted in comprehending causal relationships and mechanisms, often requiring specific domain knowledge and expertise.

**Advantages:**
- Uncovers fundamental processes and causal relationships
- Produces more interpretable and generalizable models
- Informs future interventions and strategies by identifying key factors and their relationships

**Disadvantages:**
- Requires extensive domain-specific expertise
- Can be computationally intensive and require continuous refinement
- Potential for overfitting or confirmation bias if based on limited data or too many assumptions

**Examples:**
- Biological models of disease progression
- Social models of group mechanics and interactions
- Economic models of market behavior

### 2. Predictive Modeling using Machine Learning

This approach leverages machine learning algorithms to predict outcomes based on input features, primarily aiming to exploit large datasets for accurate predictions or classifications without necessarily understanding the basic mechanisms.

**Advantages:**
- Less reliant on domain-specific expertise, enhancing scalability and accessibility
- Capable of high accuracy, especially with large and complex datasets
- Automates feature selection and parameter tuning, reducing human errors and biases

**Disadvantages:**
- Often lacks interpretability, potentially limiting applicability across contexts
- Susceptible to overfitting and capturing spurious correlations
- Raises ethical and societal concerns when used for decision-making without theoretical grounding

**Examples:**
- Email spam filtering
- Credit scoring in financial industries
- Image recognition for disease diagnosis

### 3. Leveraging Machine Learning to Understand Fundamental Mechanisms

This emerging approach combines the strengths of both previous methods, using machine learning to gain insights into fundamental mechanisms while also predicting outcomes. It employs advanced tools like SHAP (SHapley Additive exPlanations) values to interpret complex ML models and enhance understanding of variable relationships.

**Advantages:**
- Assesses individual contributions of features towards predictions
- Facilitates interpretation of complex "black box" models
- Improves model validation and evaluation by identifying influential features and their interactions

**Disadvantages:**
- Interpretations may still be limited for high-dimensional and nonlinear problems
- Some methods, like SHAP, can be computationally demanding
- Approximations and heuristics may introduce biases and assumptions

**Examples:**
- Healthcare: Predicting disease likelihood while understanding influential factors
- Customer churn analysis: Predicting and understanding reasons for customer attrition
- Natural Language Processing: Interpreting sentiment analysis models to understand word influence

> **Note:** For an in-depth example of this approach, refer to [this paper](https://www.nature.com/articles/s41598-023-34543-8) on using SHAP values to understand mechanisms of unconventional food plant consumption in Brazil.

> **Interesting Application:** SHAP values have been used to assess ML models' usefulness in predicting hypoxaemia during surgery, outperforming anesthesiologists. [Read more here](https://rdcu.be/baVbR).

## Understanding Supervised Learning

Supervised learning is a fundamental concept in machine learning where algorithms learn to make predictions or decisions based on labeled training data. This approach involves mapping input features to known output labels to create a model that can generalize to new, unseen data.

### Key Components and Goals

1. **Training Data:**
    - Consists of input features (independent variables) and corresponding output labels (dependent variables)
    - Used to approximate the relationship between inputs and outputs

2. **Model or Hypothesis:**
    - A function that maps input features to predicted outputs
    - Trained using the provided data to find optimal parameters

3. **Goal:**
    - Learn a function $\mathcal{f}: \mathcal{X} \mapsto \mathcal{Y}$, where:
        - $\mathcal{X}$ is the input space (feature set)
        - $\mathcal{Y}$ is the output space (label set)
        - $\mathcal{f}$ is the mapping function

4. **Dataset Representation:**
    - Given as $\mathcal{D} = \{(x_1, y_1), ..., (x_m, y_m)\}$
    - Where $x_i \in \mathcal{X}$ and $y_i \in \mathcal{Y}$

### Types of Supervised Learning Problems

1. **Binary Classification:**
    - Output space $\mathcal{Y} = \{\text{No}, \text{Yes}\}$ or $\{0, 1\}$
    - Predicts one of two possible classes

2. **Multiclass or Multilabel Classification:**
    - Output space $\mathcal{Y} = \{Class_1, Class_2, ..., Class_n\}$
    - Predicts one (multiclass) or multiple (multilabel) classes from several options

3. **Regression:**
    - Output space $\mathcal{Y} = \mathbb{R}$ (real numbers)
    - Predicts a continuous value

### The Learning Process

1. **Model Training:**
    - Algorithm learns from labeled training data
    - Adjusts model parameters to minimize prediction errors

2. **Generalization:**
    - The trained model should perform well on new, unseen data
    - Balancing between fitting training data and avoiding overfitting

3. **Evaluation:**
    - Assess model performance using separate test data
    - Metrics vary based on the problem type (e.g., accuracy for classification, mean squared error for regression)

### Challenges and Considerations

- **Data Quality:** The model's performance heavily depends on the quality and representativeness of the training data.
- **Feature Selection:** Choosing relevant features is crucial for model effectiveness and efficiency.
- **Model Complexity:** Balancing between model simplicity and ability to capture complex patterns in the data.
- **Overfitting vs. Underfitting:** Ensuring the model generalizes well without memorizing training data or failing to capture important patterns.
- **Hyperparameter Tuning:** Optimizing model parameters to improve performance and avoid overfitting.


```python
# Import required libraries for generating and plotting classification data
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")  # Define a custom style for matplotlib

from sklearn.datasets import make_classification

# Generate a fake classification dataset using scikit-learn's make_classification function
X, y = make_classification(
    n_samples=200,
    n_features=2,  # Use 100 samples with 2 features
    n_redundant=0,
    n_informative=2,  # Considering that no features are redundant and both are informative
    random_state=271828,
)  # Set the random state for reproducible results
X[:10], y[:10]
```

```python
# Create a scatter plot of the generated data points with different colors based on their class label
plt.scatter(
    X[y == 0, 0], X[y == 0, 1], marker="o", c="b", s=50, edgecolor="k", label="class 0"
)
plt.scatter(
    X[y == 1, 0], X[y == 1, 1], marker="o", c="r", s=50, edgecolor="k", label="class 1"
)

# Set x limits
plt.xlim(min(X[:, 0]) * 1.1, max(X[:, 0]) * 1.1)

# Set y limits
plt.ylim(min(X[:, 1]) * 1.1, max(X[:, 1]) * 1.1)

# Add labels
plt.xlabel("feature 1")
plt.ylabel("feature 2")

# Add legend - Use the classes from y
plt.legend()

# Display the plot
plt.show()
```

When a model "learns" it is actually adjusting its internal parameters (weights and biases) so that a given input leads to the desired output. It's slicing the input space in a way that the output space is divided in the desired way. The model is learning the decision boundary that separates the classes.
In the following cell, we'll train a Logistic Regression model and plot its decision boundary.

```python
from sklearn.linear_model import LogisticRegression
import numpy as np


# Instantiate a logistic regression model with a fixed random state for reproducibility
clf = LogisticRegression(random_state=271828)

# Fit the logistic regression model to the given data (X, y)
clf.fit(X, y)

# Create a grid to evaluate the model
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = 0.02  # Step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Obtain the predictions for each point in the grid
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary and fill the background colors
plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], marker="o", c=y, s=25, edgecolor="k")

# Set x limits
plt.xlim(min(X[:, 0]) * 1.1, max(X[:, 0]) * 1.1)

# Set y limits
plt.ylim(min(X[:, 1]) * 1.1, max(X[:, 1]) * 1.1)

plt.show()
```

```python
# Calculate the accuracy for the decision boundary
accuracy = clf.score(X, y)
print(f"Accuracy: {accuracy*100:.2f} %")
```

## Understanding Unsupervised Learning

Unsupervised learning is a fundamental branch of machine learning where algorithms discover patterns, structures, or relationships in data without labeled training examples. This approach is particularly valuable for exploratory data analysis, pattern recognition, and preprocessing tasks.

### Key Components and Goals

1. **Unlabeled Data:**
    - Consists of input features only, without corresponding output labels
    - Used to uncover innate structures or relationships within the dataset

2. **Learning Objective:**
    - Identify patterns, groupings, or representations in the data
    - Transform or reduce the data to reveal its basic structure

3. **Goal:**
    - Learn a function $g: \mathcal{X} \mapsto \mathcal{Z}$, where:
        - $\mathcal{X}$ is the input space (feature set)
        - $\mathcal{Z}$ is the transformed or reduced output space
        - $\mathcal{g}$ is the mapping function that reveals hidden structures

4. **Dataset Representation:**
    - Given as $\mathcal{D} = \{x_1, ..., x_m\}$
        - Where $x_i \in \mathcal{X}$

### Types of Unsupervised Learning Tasks

1. **Clustering:**
    - Groups similar data points together based on innate similarities
    - Examples: K-means, hierarchical clustering, DBSCAN

2. **Dimensionality Reduction:**
    - Reduces the number of features while preserving important information
    - Examples: Principal Component Analysis (PCA), t-SNE, autoencoders

3. **Anomaly Detection:**
    - Identifies unusual patterns or outliers in the data
    - Examples: Isolation Forest, One-Class SVM

4. **Association Rule Learning:**
    - Discovers interesting relationships between variables in large datasets
    - Example: Apriori algorithm for market basket analysis (e.g., "People who buy A also buy B")

### The Learning Process

1. **Data Exploration:**
    - Analyze the structure and distribution of the unlabeled data

2. **Pattern Recognition:**
    - Apply algorithms to identify innate groupings or relationships

3. **Model Evaluation:**
    - Assess the quality of discovered patterns using internal validation metrics
    - Examples: Silhouette score for clustering, reconstruction error for autoencoders

4. **Interpretation:**
    - Analyze and interpret the discovered patterns or structures
    - Often requires domain expertise to derive meaningful insights

### Advantages and Challenges

**Advantages:**
- Discovers hidden patterns without the need for labeled data
- Useful for exploratory data analysis and hypothesis generation
- Can handle large volumes of multidimensional data efficiently

**Challenges:**
- Results can be subjective and require careful interpretation
- Difficult to evaluate the quality of results without ground truth
- May uncover spurious patterns in noisy or unstructured data

### Applications

- Customer segmentation in marketing
- Topic modeling in natural language processing
- Anomaly detection in time-series data
- Image compression and feature extraction

### Relationship to Supervised Learning

Unsupervised learning often complements supervised learning:
- It can be used for feature extraction or data preprocessing before applying supervised methods
- Clustering can create pseudo-labels for semi-supervised learning approaches
- Dimensionality reduction can improve the efficiency and performance of supervised models


Unlike supervised learning, unsupervised learning algorithms aim to reveal hidden structures within the data without trying to predict specific outputs based on labeled examples. This makes unsupervised learning techniques particularly useful for exploratory data analysis, pattern recognition, and preprocessing for supervised learning tasks.

```python
# Create some fake data for clustering

# Import required libraries for generating and plotting clustering data
from sklearn.datasets import make_blobs

# Generate a fake clustering dataset using scikit-learn's make_blobs function
X, y = make_blobs(
    n_samples=300,
    n_features=10,  # Use 100 samples with 2 features
    centers=3,
    cluster_std=5.0,  # Generate 3 clusters with a standard deviation of 0.5
    random_state=271828,
)  # Set the random state for reproducible results

X[:10], y[:10]
```

```python
# We cannot plot data with 5 dimensions, so we will use something to reduce the dimensionality of the data to 2. Let's use PCA for that.

# Import the PCA class from scikit-learn
from sklearn.decomposition import PCA

# Instantiate a PCA object with 2 components
pca = PCA(n_components=2)

# Fit the PCA object to the data
pca.fit(X)

# Transform the data using the PCA object
X_pca = pca.transform(X)

# Plot the data points in the new 2D space
plt.scatter(X_pca[:, 0], X_pca[:, 1], marker="o", s=25, edgecolor="k")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA transformed data")
plt.show()
```

```python
# We can also use other dimensionality reduction techniques, such as t-SNE (t-Distributed Stochastic Neighbor Embedding) and SVD (Singular Value Decomposition) to reduce the dimensionality of our data.

# Import the t-SME class from scikit-learn
from sklearn.manifold import TSNE

# Instantiate a TSNE object with 2 components
tsne = TSNE(n_components=2)

# Fit the tsne object to the data and transform
X_tsne = tsne.fit_transform(X)

# Plot the data points in the new 2D space
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], marker="o", s=25, edgecolor="k")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("TSNE transformed data")
plt.show()
```

```python
# Import the SVD class from scikit-learn
from sklearn.decomposition import TruncatedSVD

# Instantiate a SVD object with 2 components
svd = TruncatedSVD(n_components=2)

# Fit the svd object to the data and transform
X_svd = svd.fit_transform(X)

# Plot the data points in the new 2D space
plt.scatter(X_svd[:, 0], X_svd[:, 1], marker="o", s=25, edgecolor="k")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("SVD transformed data")
plt.show()
```

```python
# Apply k-means clustering to the data

# Load the KMeans class from scikit-learn
from sklearn.cluster import KMeans

# Instantiate a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the KMeans object to the data
kmeans.fit(X_pca)

# Get the cluster labels
labels = kmeans.labels_

# Plot the data points in the new 2D space with the cluster labels
plt.scatter(X_pca[:, 0], X_pca[:, 1], marker="o", c=labels, s=25, edgecolor="k")

# Plot the centroids of the clusters
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker="x",
    s=100,
    c="r",
    edgecolor="k",
    label="Centroids",
)

plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA transformed data with k-means clustering")
plt.legend()
plt.show()
```

```python
# Instantiate a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the KMeans object to the data
kmeans.fit(X_tsne)

# Get the cluster labels
labels = kmeans.labels_

# Plot the data points in the new 2D space with the cluster labels
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], marker="o", c=labels, s=25, edgecolor="k")

# Plot the centroids of the clusters
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    marker="x",
    s=100,
    c="r",
    edgecolor="k",
    label="Centroids",
)

plt.xlabel("First Component")
plt.ylabel("Second Component")
plt.title("t-SNE transformed data with k-means clustering")
plt.legend()
plt.show()
```

## Data Splitting in Machine Learning

Data splitting is a crucial practice in machine learning that involves dividing a dataset into distinct subsets for training, validation, and testing purposes. This process is fundamental for developing robust models, preventing overfitting, and accurately assessing performance.

### Why Split the Data?

1. **Training Set**: This subset is used to teach the model, allowing it to learn patterns and relationships between input features and target outputs.

2. **Validation Set**: Used for fine-tuning the model and selecting the best configuration. It plays a critical role in hyperparameter optimization and model selection.

3. **Test Set**: Provides an unbiased evaluation of the final model's performance on unseen data. It's crucial to use this set only once to avoid inadvertently fitting the model to the test data.

### Benefits of Data Splitting

- **Overfitting Prevention**: By monitoring performance on separate validation data, we can detect and mitigate overfitting to the training set.
- **Effective Model Selection**: The validation set allows for comparison of different model architectures and hyperparameters without compromising the integrity of the test set.
- **Realistic Performance Estimation**: An independent test set provides a more accurate representation of how the model will perform on new, unseen data in real-world applications.

> **Important Note**: Always use pseudo-random number generators with fixed seeds for reproducibility in your experiments. This ensures consistent results across multiple runs while maintaining the benefits of randomization.


### Common Data Splitting Methods

#### Holdout Method

The holdout method is the most common method for splitting the data. It involves randomly splitting the data into two sets: training and test. The training set is used to train the model, while the test set is used to evaluate the model's performance. The test set is only used once, at the end of the training process, to estimate the model's performance on previously unseen data. This approach is shown in the figure below.

### K-fold Cross Validation

K-fold cross-validation involves dividing the dataset into K equal-sized folds. The process repeats K times, where each time one of the K folds is used as the validation set while the remaining K-1 folds are used to train the model. The average performance metric across all K iterations provides the final model evaluation result.

#### Steps for K-fold Cross Validation

1. **Shuffle the dataset**: Randomly order the data to ensure an unbiased distribution across all folds.
2. **Divide the dataset**: Partition the dataset into K equally sized folds.
3. **Iterate through the folds**: For each iteration, use one fold as the validation set and the remaining K-1 folds as the training set.
4. **Train and validate the model**: Fit the model on the current training set and evaluate its performance on the validation set.
5. **Average the performance metrics**: After completing all iterations, calculate the average performance metric (e.g., accuracy, F1-score) across all the K validation sets.

#### Advantages of K-fold Cross Validation

- **Reduced Variability**: The averaging of performance metrics across multiple validation sets reduces variability and prevents over-reliance on any single partitioning scheme.
- **Better Employment of Data**: In comparison to a fixed train-validation-test split, K-fold cross-validation enables the model to be trained and validated using the entire dataset. This leads to more extensive training and performance evaluation.
- **More Accurate Model Selection & Hyperparameter Tuning**: Evaluating models with K-fold cross-validation provides a better understanding of how well they generalize. This is particularly beneficial for model comparison and hyperparameter tuning.

#### Choosing the Value of K

The choice of K depends on the specific problem, dataset size, and computation resources available. Some common choices are:

- **K = 5**: Provides a good balance between accuracy and computation time, as long as there's enough data.
- **K = 10**: Often considered the standard choice; it offers a more accurate estimate but may require more time to compute.
- **K = 20**: Provides a more accurate estimate but may be too computationally expensive.

Keep in mind that larger values of K result in smaller differences between training and validation sets, potentially leading to a better estimation of model performance. However, this advantage comes at the cost of increased computation time.

```python
from sklearn.model_selection import KFold

# Instantiate a KFold object with 5 splits
kf = KFold(n_splits=5, shuffle=True, random_state=271828)

# Iterate over the splits and print the number of the folder and the sizes of the training and validation sets
for i, (train_index, val_index) in enumerate(kf.split(X)):
    print(f"Fold {i+1}")
    print(f"Training set: {train_index.shape} - idx: {train_index[:5]}")
    print(f"Validation set: {val_index.shape} - idx: {val_index[:5]}")
    print()
```

### Special Cases in Data Splitting

#### Imbalanced Classes

When dealing with imbalanced classes, it is essential to ensure that the training and test sets contain an equal proportion of samples from each class. Stratified sampling can achieve this by maintaining the same proportion of samples from each class in both sets.



#### Time Series
When working with time series data, it is crucial to preserve the temporal order of the data. This means that the data should not be randomly shuffled before splitting. Instead, the data should be split in chronological order, with the oldest data points used for training and the most recent data points used for testing.

The order of the data points is important because each data point is dependent on the previous ones. For example, in stock market data, the price of a stock today is likely to be influenced by the price of the stock yesterday. If we randomly shuffle the data before splitting it into training and testing sets, we risk including future data in the training set, which can lead to overfitting and poor generalization performance. By splitting the data in chronological order, we ensure that the model is trained on past data and tested on future data, which is a more realistic scenario of what it will encounter in production.

One example of a medical application is electrocardiogram (ECG) data. ECG data is a time series of electrical signals that are recorded from the heart. The signals can be used to diagnose various heart conditions, such as arrhythmias and heart attacks. When training a model to diagnose heart conditions using ECG data, it is important to ensure that the training and test sets are temporally disjoint. This means that the training set should contain ECG data from a time period before the test set, and vice versa. Ensuring temporal disjointness prevents the model from learning patterns specific to the test set, which can lead to overfitting and poor generalization performance.

A common approach is to create a "window" of data, using the data within the window as the training set and the data following the window as the test set. The window then moves forward, and the process is repeated. This technique is known as walk-forward validation. The window can either be a fixed size rolling through time or an ever-expanding window, referred to as expanding window validation.

When handling time-sensitive data, it's important to make sure that the training and test sets are temporally disjoint. This means that the training set should have data from a time period before the test set, and vice versa. Ensuring temporal disjointness prevents the model from learning patterns specific to the test set. 4. Data splitting is used to divide a dataset into distinct subsets for training, validation, and testing purposes, helping to prevent overfitting and accurately assess model performance.
